```python
deck_list = list()
for cabin in titanic_df.Cabin:
    if type(cabin) is str:
        deck_list.append(cabin[0])
    else:
        deck_list.append(np.nan)
```

```python
titanic_df['Deck'] = deck_list
titanic_df[(titanic_df['Deck'] == 'A') & (titanic_df['Survived'] == 1)]
```

```python
children_df = titanic_df[titanic_df["Age"] <= 18.0]
print(len(children_df[children_df["Survived"] == 1]) / len(children_df))
```

```python
adult_df = titanic_df[(titanic_df["Age"] > 18.0) & (titanic_df["Age"] <= 50)]
print(len(adult_df[adult_df["Survived"] == 1]) / len(adult_df))
```
